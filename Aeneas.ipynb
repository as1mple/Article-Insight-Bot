```python
import aeneas.globalconstants as gc
import aeneas.executetask as et
import aeneas.task as task
```

```python
audio_file_path = "1931.wav"
text_file_path = "test.txt"
output_syncmap_path = "res.txt"
```

```python
config_string = u"task_language=eng|is_text_type=plain|os_task_file_format=txt"
aeneas_task = task.Task(config_string=config_string)
aeneas_task.audio_file_path_absolute = audio_file_path
aeneas_task.text_file_path_absolute = text_file_path
aeneas_task.sync_map_file_path_absolute = output_syncmap_path
```

```python
executor = et.ExecuteTask(aeneas_task)
executor.execute()
```

```python
aeneas_task.output_sync_map_file()

print("Forced alignment is complete. Check the output at:", output_syncmap_path)
```
